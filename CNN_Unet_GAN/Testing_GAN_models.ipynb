```python
%matplotlib inline

import sys
import numpy as np
import matplotlib.pyplot as plt

from PIL import Image
import os

from keras.datasets import mnist
from keras.layers import Input, Dense, Reshape, Flatten, Dropout, Conv2D, MaxPooling2D, UpSampling2D, ZeroPadding2D
from keras.layers import BatchNormalization, Activation
from keras.layers.advanced_activations import LeakyReLU
from keras.models import Sequential, Model
from keras.optimizers import Adam
```

## First let us build the dataset from the graystacks

```python
# Import first graystack
im = Image.open('../graystacks/graystack00000000.tif')
imarray = np.array(im)
gs = imarray[26:, 26:]
gs = np.expand_dims(gs, axis=2)
# Then do that for all graystacks
list_gs = os.listdir('../graystacks')
ngs = len(list_gs)
for file in range(1, ngs):
    im = Image.open('../graystacks/' + list_gs[file])
    imarray = np.array(im)
    gs = np.concatenate((gs, np.expand_dims(imarray[26:, 26:], axis=2)), axis=2)
# Recover shape of one graystack
xgs, ygs, _ = gs.shape
   
print('Size of graystacks concatenate is', gs.shape)
randgs = np.random.choice(np.array(ngs))
print('Random graystack from concatenate ( layer', randgs, '):')
plt.imshow(gs[:, :, randgs])
```

```python
# Build dataset
nsize = 31
xpick = np.array(xgs-nsize)
ypick = np.array(ygs-nsize)
ntrain = 60000
X_train = np.zeros((ntrain, nsize, nsize))
for ex in range(ntrain):
    xex = np.random.choice(xpick)
    yex = np.random.choice(ypick)
    zex = np.random.choice(np.array(ngs))
    X_train[ex, :, :] = gs[xex:xex+nsize, yex:yex+nsize, zex]
    
randds = np.random.choice(np.array(ntrain))
print('Random example from dataset ( example', randds, '):')
plt.imshow(X_train[randds, :, :])
```

## Now building GAN class, using the following website
https://towardsdatascience.com/writing-your-first-generative-adversarial-network-with-keras-2d16fd8d4889

```python
class GAN():
    def __init__(self, img_rows, img_cols, channels):
        self.img_rows = img_rows
        self.img_cols = img_cols
        self.channels = channels
        self.img_shape = (self.img_rows, self.img_cols, self.channels)
        self.latent_dim = 100 # this will the shape of the input noise provided to the generator
        # Define optimizer
        optimizer = Adam(0.0002, 0.5)      
        # Define discriminator and compile it, which will be trained on its side
        self.discriminator = self.build_discriminator() 
        self.discriminator.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy']) 
        # Define generator
        self.generator = self.build_generator()
        # Define input to generator z and output of generator img
        z = Input(shape=(self.latent_dim,))
        img = self.generator(z)
        # When training the generator we need the output of the total system, 
        # but we do not want to train the discriminator, so we freeze its weights
        self.discriminator.trainable = False
        # Define output of the discriminator
        validity = self.discriminator(img)
        # Define full model to be trained and compile it
        self.combined = Model(z, validity)
        self.combined.compile(loss='binary_crossentropy', optimizer=optimizer)
        
    def build_generator(self):
        # Define model as sequential
        model = Sequential()
        # Add successive layers to model
        model.add(Dense(256, input_dim=self.latent_dim))
        model.add(LeakyReLU(alpha=0.2))
        model.add(BatchNormalization(momentum=0.8))
        model.add(Dense(512))
        model.add(LeakyReLU(alpha=0.2))
        model.add(BatchNormalization(momentum=0.8))
        model.add(Dense(1024))
        model.add(LeakyReLU(alpha=0.2))
        model.add(BatchNormalization(momentum=0.8))
        model.add(Dense(np.prod(self.img_shape), activation='tanh'))
        model.add(Reshape(self.img_shape))
        # Summary of the model
        model.summary()
        # Input is the noise vector 
        noise = Input(shape=(self.latent_dim,))
        # Output after going through the model (an image)
        img = model(noise)
        # And finally return the Model
        return Model(noise, img)
    
    def build_discriminator(self):
        # Define model as sequential
        model = Sequential()
        # Add successive layers to model
        model.add(Flatten(input_shape=self.img_shape))
        model.add(Dense(512))
        model.add(LeakyReLU(alpha=0.2))
        model.add(Dense(256))
        model.add(LeakyReLU(alpha=0.2))
        model.add(Dense(1, activation='sigmoid'))
        # Summary of the model
        model.summary()
        # Input is the image 
        img = Input(shape=self.img_shape)
        # Output after going through the model (a number)
        validity = model(img)
        # And finally return the Model
        return Model(img, validity)
    
    def train(self, X_train, epochs, batch_size=128, sample_interval=50):        
        # Adjust dataset
        self.min_dataset = np.min(X_train)
        self.max_dataset = np.max(X_train)
        X_train = (X_train-self.min_dataset) / (0.5*(self.max_dataset-self.min_dataset)) - 1.
        if len(X_train.shape) < 4:
            X_train = np.expand_dims(X_train, axis=3) 
        # Create labels
        valid = np.ones((batch_size, 1))
        fake = np.zeros((batch_size, 1))
        # Then train on batch at each epoch the two models
        for epoch in range(epochs):   
            # Get examples from data set X_train
            idx = np.random.randint(0, X_train.shape[0], batch_size)
            imgs = X_train[idx]  
            # Build input noise for generator
            noise = np.random.normal(0, 1, (batch_size, self.latent_dim))    
            # Get output of generator
            gen_imgs = self.generator.predict(noise)   
            # Train discriminator based on batch and output of the generator
            d_loss_real = self.discriminator.train_on_batch(imgs, valid)
            d_loss_fake = self.discriminator.train_on_batch(gen_imgs, fake)
            d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)
            # Train combined model: if generator works well, discriminator will think it came from actual dataset
            noise = np.random.normal(0, 1, (batch_size, self.latent_dim))
            g_loss = self.combined.train_on_batch(noise, valid) # we want the output to always be 1, hence valid
            # Print indication
            print ("%d [D loss: %f, acc.: %.2f%%] [G loss: %f]" % (epoch, d_loss[0], 100*d_loss[1], g_loss))
            if epoch % sample_interval == 0:
                self.sample_images(epoch)
                
    def sample_images(self, epoch):
        r, c = 3, 3
        noise = np.random.normal(0, 1, (r * c, self.latent_dim))
        gen_imgs = self.generator.predict(noise)        
        gen_imgs = 0.5 * gen_imgs + 0.5
        fig, axs = plt.subplots(r, c)
        cnt = 0
        for i in range(r):
            for j in range(c):
                axs[i,j].imshow(gen_imgs[cnt, :,:,0], cmap='gray')
                axs[i,j].axis('off')
                cnt += 1
        fig.savefig("GAN_images/%d.png" % epoch)
        plt.close()
```

There might be the following error message <br/><br/>

'Discrepancy between trainable weights and collected trainable' <br/><br/>
According to me, this is due to the fact that we compile the discriminator and then freeze its weights for combined model. Python thinks we should freeze before compiling when in fact, we actually mean what we did.

### Remark 2)
When we train on the graystack dataset, it only provides noise, even after lots of training. I believe this is due to the fact lots of parts in the graystack are basically background with noise. So maybe in the training dataset there are many examples coming from these regions along with actual neurons. Since it might be easier to provide random data for the generator, and in the end the discriminator does not make a difference, I believe this is what happens.

# # Making new GAN with better networks
from https://towardsdatascience.com/generating-modern-arts-using-generative-adversarial-network-gan-on-spell-39f67f83c7b4

```python
class GAN():
    def __init__(self, img_rows, img_cols, channels):
        self.img_rows = img_rows
        self.img_cols = img_cols
        self.channels = channels
        self.img_shape = (self.img_rows, self.img_cols, self.channels)
        self.latent_dim = 100 # this will the shape of the input noise provided to the generator
        # Define optimizer
        optimizer = Adam(0.0002, 0.5)      
        # Define discriminator and compile it, which will be trained on its side
        self.discriminator = self.build_discriminator() 
        self.discriminator.compile(loss='binary_crossentropy', optimizer=optimizer, metrics=['accuracy']) 
        # Define generator
        self.generator = self.build_generator(2)
        # Define input to generator z and output of generator img
        z = Input(shape=(self.latent_dim,))
        img = self.generator(z)
        # When training the generator we need the output of the total system, 
        # but we do not want to train the discriminator, so we freeze its weights
        self.discriminator.trainable = False
        # Define output of the discriminator
        validity = self.discriminator(img)
        # Define full model to be trained and compile it
        self.combined = Model(z, validity)
        self.combined.compile(loss='binary_crossentropy', optimizer=optimizer)
        
    def build_generator(self, generate_res):
        # New model
        model = Sequential()
        model.add(Dense(4 * 4 * 256, activation='relu', input_dim=self.latent_dim))
        model.add(Reshape((4, 4, 256)))    
        model.add(UpSampling2D())
        model.add(Conv2D(256, kernel_size=3, padding='same'))
        model.add(BatchNormalization(momentum=0.8))
        model.add(Activation('relu'))    
        model.add(UpSampling2D())
        model.add(Conv2D(256, kernel_size=3, padding='same'))
        model.add(BatchNormalization(momentum=0.8))
        model.add(Activation('relu'))    
        for i in range(generate_res):
            model.add(UpSampling2D())
            model.add(Conv2D(256, kernel_size=3, padding='same'))
            model.add(BatchNormalization(momentum=0.8))
            model.add(Activation('relu')) 
        model.add(Conv2D(self.channels, kernel_size=3, padding='same'))
        model.add(Activation('tanh'))
        # Summary of the model
        model.summary()
        # Input is the noise vector 
        noise = Input(shape=(self.latent_dim,))
        # Output after going through the model (an image)
        img = model(noise)
        # And finally return the Model
        return Model(noise, img)
    
    def build_discriminator(self):
        # New model
        model = Sequential()    
        model.add(Conv2D(32, kernel_size=3, strides=2, input_shape=self.img_shape, padding='same'))
        model.add(LeakyReLU(alpha=0.2))
        model.add(Dropout(0.25))    
        model.add(Conv2D(64, kernel_size=3, strides=2, padding='same'))
        model.add(ZeroPadding2D(padding=((0, 1), (0, 1))))
        model.add(BatchNormalization(momentum=0.8))
        model.add(LeakyReLU(alpha=0.2))
        model.add(Dropout(0.25))    
        model.add(Conv2D(128, kernel_size=3, strides=2, padding='same'))
        model.add(BatchNormalization(momentum=0.8))
        model.add(LeakyReLU(alpha=0.2))
        model.add(Dropout(0.25))    
        model.add(Conv2D(256, kernel_size=3, strides=1, padding='same'))
        model.add(BatchNormalization(momentum=0.8))
        model.add(LeakyReLU(alpha=0.2))    
        model.add(Dropout(0.25))
        model.add(Conv2D(512, kernel_size=3, strides=1, padding='same'))
        model.add(BatchNormalization(momentum=0.8))
        model.add(LeakyReLU(alpha=0.2))    
        model.add(Dropout(0.25))
        model.add(Flatten())
        model.add(Dense(1, activation='sigmoid'))
        # Summary of the model
        model.summary()
        # Input is the image 
        img = Input(shape=self.img_shape)
        # Output after going through the model (a number)
        validity = model(img)
        # And finally return the Model
        return Model(img, validity)
    
    def train(self, X_train, epochs, batch_size=128, sample_interval=50):        
        # Adjust dataset
        self.min_dataset = np.min(X_train)
        self.max_dataset = np.max(X_train)
        X_train = (X_train-self.min_dataset) / (0.5*(self.max_dataset-self.min_dataset)) - 1.
        if len(X_train.shape) < 4:
            X_train = np.expand_dims(X_train, axis=3) 
        # Create labels
        valid = np.ones((batch_size, 1))
        fake = np.zeros((batch_size, 1))
        # Then train on batch at each epoch the two models
        for epoch in range(epochs):   
            # Get examples from data set X_train
            idx = np.random.randint(0, X_train.shape[0], batch_size)
            imgs = X_train[idx]  
            # Build input noise for generator
            noise = np.random.normal(0, 1, (batch_size, self.latent_dim))    
            # Get output of generator
            gen_imgs = self.generator.predict(noise)   
            # Train discriminator based on batch and output of the generator
            d_loss_real = self.discriminator.train_on_batch(imgs, valid)
            d_loss_fake = self.discriminator.train_on_batch(gen_imgs, fake)
            d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)
            # Train combined model: if generator works well, discriminator will think it came from actual dataset
            noise = np.random.normal(0, 1, (batch_size, self.latent_dim))
            g_loss = self.combined.train_on_batch(noise, valid) # we want the output to always be 1, hence valid
            # Print indication
            print ("%d [D loss: %f, acc.: %.2f%%] [G loss: %f]" % (epoch, d_loss[0], 100*d_loss[1], g_loss))
            if epoch % sample_interval == 0:
                self.sample_images(epoch)
                
    def sample_images(self, epoch):
        r, c = 3, 3
        noise = np.random.normal(0, 1, (r * c, self.latent_dim))
        gen_imgs = self.generator.predict(noise)        
        gen_imgs = 0.5 * gen_imgs + 0.5
        fig, axs = plt.subplots(r, c)
        cnt = 0
        for i in range(r):
            for j in range(c):
                axs[i,j].imshow(gen_imgs[cnt, :,:,0], cmap='gray')
                axs[i,j].axis('off')
                cnt += 1
        fig.savefig("GAN_images/%d.png" % epoch)
        plt.close()
```
