```python
import time
import random
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import multivariate_normal
import tensorflow as tf
import keras
import keras.backend as K
from keras.layers import Input, Conv1D, Conv2D, MaxPooling2D, BatchNormalization, Flatten, Dense
from keras import layers
from keras.models import Model
import multiprocessing as mp
from scipy.ndimage import convolve

from CNN_functions import create_plans, stack_plans, build_dataset, compress_plans
#import CNN_functions as cnnf

%matplotlib inline
```

```python
# Common to all plans
dim_plan_pix = np.array([80, 80])
# Neurons characteristics
cov_mean = 0
cov_std = 0.15
xvar_mean = 4.5
xvar_std = 1
yvar_mean = 4.5
yvar_std = 1
# Matrix for plans
planpar = np.array([[250, 150, 100], #, 15], # value for neurons
                    [50, 30, 20], #, 4], # variance for value for neurons
                    [3, 35, 25], #, 37], # repartitions of values for neurons 
                    [60, 100, 100], #, 15], # mean number of neurons per plan
                    [3, 5, 5]]) #, 5]]) # std for number of neurons per plan

# Building first empty matrices
plan_sketch = np.zeros((0, dim_plan_pix[0], dim_plan_pix[1]))
coord_sketch = np.zeros((0, 3))

# Launching for-loop
ti = time.time()
for i in range(planpar.shape[1]):
    plannum = planpar[2, i]
    numneur = planpar[3, i]
    numstd = planpar[4, i]
    info_neuron = planpar[0, i], planpar[1, i], cov_mean, cov_std, xvar_mean, xvar_std, yvar_mean, yvar_std
    plan_sketch_temp, coord_sketch_temp = create_plans(plannum, dim_plan_pix, numneur, numstd, info_neuron)
    len_plan_sketch = plan_sketch.shape[0]
    coord_sketch_temp[:, 0] += len_plan_sketch
    plan_sketch = np.concatenate((plan_sketch, plan_sketch_temp), axis=0)
    coord_sketch = np.vstack((coord_sketch, coord_sketch_temp))
tf = time.time() - ti

print("Size of plan is ", plan_sketch.shape, ", with ", coord_sketch.shape[0], " neurons. Computed in ", tf, " seconds.")

# Example to plot
example = 0
# Plotting original neurons
plt.imshow(plan_sketch[example, :, :])
plt.colorbar()
plt.title("Plan with artificially created neurons from create_plan (detail)")
# Adding centers of neurons
coord_sketch2 = coord_sketch[coord_sketch[:, 0] == example, 1:3]
plt.scatter(coord_sketch2[:, 1], coord_sketch2[:, 0], s=5, c=[1, 0, 0]*np.ones((coord_sketch2.shape[0], 1)), marker='.')
```

```python
# Building plans based on plan_sketch and coord_sketch
# Parameters
nstacks = 5000
stk_mean = 8
stk_std = 3
crop_size = 30
noise_mean = 0
noise_std = 0

# Launching for-loop
ti = time.time()
sk_plans, sk_coord = stack_plans(nstacks, plan_sketch, coord_sketch, crop_size, stk_mean, stk_std, noise_mean, noise_std)
# NORMALIZING SK_PLAN
sk_plans = sk_plans / np.max(sk_plans)
tf = time.time() - ti

print("Size of sk_plans is ", sk_plans.shape, ", and there are ", sk_coord.shape[0], " neurons. Computed in ", tf, " seconds.")

# Plotting random stack with pointers for neuron centers
example = 0
# Plotting original neurons
plt.imshow(sk_plans[example, :, :])
plt.colorbar()
plt.title("Final plan with artificially created neurons using stack_plans (detail)")
# Adding centers of neurons
sk_coord2 = sk_coord[sk_coord[:, 0] == example, 1:3]
plt.scatter(sk_coord2[:, 1], sk_coord2[:, 0], s=5, c=[1, 0, 0]*np.ones((sk_coord2.shape[0], 1)), marker='.')
```

```python
# Buildind dataset
training_proportion = 0.8
ti = time.time()
Xtrain, Ytrain, Xtest, Ytest = build_dataset(sk_plans, sk_coord, training_proportion)
tf = time.time() - ti

print("Size of training set is ", Xtrain.shape, Ytrain.shape, ", test set is ", Xtest.shape, Ytest.shape, ". Computed in ", tf, " seconds.")

# Rearranging to CNN architecture
# Reshaping
Xtrain = Xtrain.reshape(Xtrain.shape[0], crop_size, crop_size, 1)
#Ytrain = Ytrain.reshape(Ytrain.shape[0], 1, 1, crop_size*crop_size)
Ytrain = np.sum(Ytrain, axis=1).reshape(Ytrain.shape[0], 1, 1, 1)
Xtest = Xtest.reshape(Xtest.shape[0], crop_size, crop_size, 1)
#Ytest = Ytest.reshape(Ytest.shape[0], 1, 1, crop_size*crop_size)
Ytest = np.sum(Ytest, axis=1).reshape(Ytest.shape[0], 1, 1, 1)

print("New sizes for training set are ", Xtrain.shape, Ytrain.shape, " test set are ", Xtest.shape, Ytest.shape)
```

Using function compress_plans from CNN_functions.py, we can compress dataset if needed.

```python
# Testing binary crossentropy keras functionality with 1 output, with new labels
# describing if there is more or less than a number of neurons for each plan.
# Y labels have 2 components
# Works with Flatten, now we are trying CNN until the end
Ytr = np.sum(Ytrain.reshape(Ytrain.shape[0], crop_size*crop_size), axis=1)
mean_neu = np.mean(Ytr)
Ytr1 = 1 * (Ytr > mean_neu-10)
Ytr2 = 1 * (Ytr > mean_neu+10)
Ytr = np.hstack((Ytr1[:, np.newaxis], Ytr2[:, np.newaxis]))
Yte = np.sum(Ytest.reshape(Ytest.shape[0], crop_size*crop_size), axis=1)
Yte1 = 1 * (Yte > mean_neu-10)
Yte2 = 1 * (Yte > mean_neu+10)
Yte = np.hstack((Yte1[:, np.newaxis], Yte2[:, np.newaxis]))
print("Mean number of neurons is ", mean_neu)
# Keras model
sk_in = Input(shape=(crop_size, crop_size, 1))
x = Conv2D(32, (3, 3), strides=(1, 1), padding="same", activation="relu", name="1st_conv")(sk_in)
x = Conv2D(32, (3, 3), strides=(1, 1), padding="same", activation="relu", name="2nd_conv")(x)
x = BatchNormalization(name="1st_norm")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="1st_pool")(x)
x = Conv2D(64, (3, 3), strides=(1, 1), padding="same", activation="relu", name="3rd_conv")(x)
x = Conv2D(64, (3, 3), strides=(1, 1), padding="same", activation="relu", name="4th_conv")(x)
x = Conv2D(64, (3, 3), strides=(1, 1), padding="same", activation="relu", name="5th_conv")(x)
x = BatchNormalization(name="2nd_norm")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="2nd_pool")(x)
x = Conv2D(128, (3, 3), strides=(1, 1), padding="same", activation="relu", name="6th_conv")(x)
x = Conv2D(128, (3, 3), strides=(1, 1), padding="same", activation="relu", name="7th_conv")(x)
x = Conv2D(128, (3, 3), strides=(1, 1), padding="same", activation="relu", name="8th_conv")(x)
x = BatchNormalization(name="3rd_norm")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="3rd_pool")(x)
#x = Flatten()(x)
#sk_out = Dense(2, activation="sigmoid", name="dense_layers")(x)
f = int(np.floor(crop_size/8))
sk_out = Conv2D(2, (f, f), strides=(1, 1), padding="valid", activation="sigmoid", name="9th_conv")(x)
sk_model = Model(inputs=sk_in, outputs=sk_out)
# Fit model
sk_model.compile(optimizer="adam", loss="binary_crossentropy", metrics=["accuracy"])
sk_model.fit(Xtrain, Ytr.reshape(Ytr.shape[0], 1, 1, Ytr.shape[1]), batch_size=32, epochs=5)
# Evaluate model
loss, accuracy = sk_model.evaluate(Xtest, Yte.reshape(Yte.shape[0], 1, 1, Yte.shape[1]), batch_size=32)
print("Accuracy on test set is ", accuracy)
#Predict for example of test set
example = 2
coord_pred = sk_model.predict(Xtrain[example, :, :, :].reshape(1, crop_size, crop_size, 1))
print("Prediction is ", coord_pred.reshape(1, Yte.shape[1]), ", and real labels are ", Yte[example, :])
```

```python
# Now testing again NN for neurons centers detection
# Keras model
sk_in = Input(shape=(crop_size, crop_size, 1))
x = Conv2D(16, (3, 3), strides=(1, 1), padding="same", activation="relu", name="1st_conv")(sk_in)
x = BatchNormalization(name="1st_norm")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="1st_pool")(x)
x = Conv2D(32, (5, 5), strides=(1, 1), padding="same", activation="relu", name="2nd_conv")(x)
x = BatchNormalization(name="2nd_norm")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="2nd_pool")(x)
x = Conv2D(64, (5, 5), strides=(1, 1), padding="same", activation="relu", name="3rd_conv")(x)
x = BatchNormalization(name="3rd_norm")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="3rd_pool")(x)
x = Conv2D(128, (5, 5), strides=(1, 1), padding="same", activation="relu", name="4th_conv")(x)
x = BatchNormalization(name="4th_norm")(x)
x = MaxPooling2D((2, 2), strides=(2, 2), padding="valid", name="4th_pool")(x)
#x = Flatten()(x)
#sk_out = Dense(2, activation="sigmoid", name="dense_layers")(x)
f = int(np.floor(crop_size/16))
x = Conv2D(crop_size*crop_size, (f, f), strides=(1, 1), padding="valid", activation="relu", name="5th_conv")(x)
x = Conv2D(crop_size*crop_size, (1, 1), strides=(1, 1), padding="valid", activation="relu", name="6th_conv")(x)
sk_out = Conv2D(1, (1, 1), strides=(1, 1), padding="valid", activation="relu", name="7th_conv")(x)
sk_model = Model(inputs=sk_in, outputs=sk_out)

# Fit model
sk_model.compile(optimizer="rmsprop", loss="mean_squared_error", metrics=["accuracy"])
sk_model.fit(Xtrain, Ytrain, batch_size=16, epochs=2)
# Evaluate model
loss, accuracy = sk_model.evaluate(Xtest, Ytest, batch_size=16)
print("Accuracy on test set is ", accuracy)

#Predict for example of test set
example = 0
coord_pred = sk_model.predict(Xtrain[example, :, :, :].reshape(1, crop_size, crop_size, 1))
# Guesses for centers of neurons with imshow
xpred, ypred = np.where(coord_pred.reshape(crop_size, crop_size) == 1)
plt.imshow(coord_pred.reshape(crop_size, crop_size).T)
plt.colorbar()
plt.title("Guess for centers of neurons")
# Adding original centers of neurons
sk_coord2 = sk_coord[sk_coord[:, 0] == example, 1:3]
plt.scatter(sk_coord2[:, 1], sk_coord2[:, 0], s=10, c=[1, 0, 0]*np.ones((sk_coord2.shape[0], 1)), marker='.')
```
